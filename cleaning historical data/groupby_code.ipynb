```python
#import csv
#import json
#import pandas as pd
country="zimbabwe"
```

```python
file_df = pd.read_json(f"{country}_fires.json", encoding="utf-8")
```

```python
file_df.head(1)
```

```python
table1 = file_df.groupby('Date').agg({'Date':'count', 'Fire intensity':'mean'})
```

```python
table1.head(1)
```

```python
new_table = table1.rename(columns={"Date":"Number of fires", "Fire intensity":"Average fire intensity"})
new_table.head(1)
```

```python
new_table.to_csv(f"{country}.csv")
```

```python
with open(f"{country}.csv") as csvr:
    reader = csv.DictReader(csvr)
    rows = list(reader)
```
