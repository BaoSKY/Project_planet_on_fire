# adding Fire Data

```python
#read fire data json file
fire_a=pd.read_json('australia.json')
fire_a
```

```python
#select columns that will be used
aus_fire=fire_a[['acq_date', 'brightness', 'frp']]
aus_fire
```

```python
#count unique values: number of active fires on a certain date
fire_count=aus_fire['acq_date'].value_counts()
fire_count
```

```python
#convert to dataframe
fire_count_df=fire_count.to_frame().reset_index()
fire_count_df
```

```python
#renaming columns
fire_count_df=fire_count_df.rename(columns={'index': 'Date',
                                            'acq_date': 'Count'
    
})
fire_count_df
```

```python
#sort starting from earlier date
fire_count_df=fire_count_df.sort_values(by='Date', ascending=True)
fire_count_df
```

```python
#groupby date and calculate mean for frp and brightness
aus_fire_group=aus_fire.groupby('acq_date').mean()
aus_fire_index=aus_fire_group.reset_index()
aus_fire_index
```

```python
aus_fire_index=aus_fire_index.rename(columns={'brightness': 'Brightness',
                                             'frp': 'FRP',
                                             'acq_date': 'Date'})
aus_fire_index
```

```python
#printing both dataframes for visibility
fire_count_df
```

```python
#printing both dataframes for visibility
aus_fire_index
```

```python
#merge means and counts
fires=pd.merge(aus_fire_index, fire_count_df, on='Date', how='outer')
fires
```

```python
#save fire data to csv
fires.to_csv('fires.csv')
```
