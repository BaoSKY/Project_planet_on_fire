# combine fire and weather data

```python
import pandas as pd
```

```python
#read fire data
fire_data=pd.read_csv('fires.csv')
fire_data
```

```python
#read weather data
weather_data=pd.read_csv('weather_data.csv')
weather_data
```

```python
#merge weather and fire data
australia=pd.merge(weather_data, fire_data, on='Date', how='left')
australia
```

```python
australia=australia[['Date', 'Temp', 'Wind', 'Precip', 'Pressure', 'Heat', 'Brightness', 'FRP', 'Count']]
australia.head(-65)
```

```python
#eliminating null values
australia=australia.fillna(0)
australia.head(-65)
```

```python
#save into csv
australia.to_csv('australia.csv')
```

# machine learning

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
aus=pd.read_csv('australia_cut2.csv')
aus
```

```python
aus_ml=pd.get_dummies(aus,prefix=['month'])
aus_ml
```

# Model 1

```python
# Assign X (data) and y (target)
X = aus_ml.drop("FRP", axis=1)
y = aus_ml["FRP"].values.reshape(-1,1)
print(X.shape, y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
regression = LinearRegression()
regression
```

```python
regression.fit(X_train_scaled, y_train)
```

```python
print(f"Training Data Score: {regression.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {regression.score(X_test_scaled, y_test)}")
```
